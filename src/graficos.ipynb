```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### GRÁFICO CINEMÁTICA DIFERENCIAL

```python
xcur = []
ycur = []
zcur = []

with open("/home/benja/proy_funda/src/kuka_iiwa_description/src/diffkine_graph/xcurrent.txt", 'r') as archivo:
    lineas = archivo.readlines()
    for linea in lineas:
        datos = linea.strip().split(' ')  # Suponiendo que los datos estén separados por comas
        xcur.append(float(datos[0]))
        ycur.append(float(datos[1]))
        zcur.append(float(datos[2]))
```

```python
xdes = []
ydes = []
zdes = []

with open("/home/benja/proy_funda/src/kuka_iiwa_description/src/diffkine_graph/xdesired.txt", 'r') as archivo:
    lineas = archivo.readlines()
    for linea in lineas:
        datos = linea.strip().split(' ')  # Suponiendo que los datos estén separados por comas
        xdes.append(float(datos[0]))
        ydes.append(float(datos[1]))
        zdes.append(float(datos[2]))
```

```python
q1 = []
q2 = []
q3 = []
q4 = []
q5 = []
q6 = []
q7 = []
q8 = []

with open("/home/benja/proy_funda/src/kuka_iiwa_description/src/diffkine_graph/q.txt", 'r') as archivo:
    lineas = archivo.readlines()
    for linea in lineas:
        datos = linea.strip().split(' ')  # Suponiendo que los datos estén separados por comas
        if len(datos) > 8:  # Verifica que haya al menos 7 elementos
            try:
                q1.append(float(datos[0]))
                q2.append(float(datos[1]))
                q3.append(float(datos[2]))
                q4.append(float(datos[3]))
                q5.append(float(datos[4]))
                q6.append(float(datos[5]))
                q7.append(float(datos[6]))
                q8.append(float(datos[7]))
            except ValueError:
                print(f"Error de conversión en la línea: {linea.strip()}")
        else:
            print(f"Línea ignorada por falta de datos: {linea.strip()}")
```

```python
t = np.arange(0, 359, 1)

plt.figure(figsize=(10,10))
# Posiciones
plt.subplot(2,3,1); plt.plot(t/50, xcur); plt.plot(t/50, xdes*np.ones(len(t)), '--')
plt.title('$x$'); plt.grid()
plt.subplot(2,3,2); plt.plot(t/50, ycur); plt.plot(t/50, ydes*np.ones(len(t)), '--')
plt.title('$y$'); plt.grid()
plt.subplot(2,3,3); plt.plot(t/50, zcur); plt.plot(t/50, zdes*np.ones(len(t)), '--')
plt.title('$z$'); plt.grid()
```

```python
t = np.arange(0, 359, 1)
plt.figure(figsize=(15,10))
# Posiciones
plt.subplot(2,4,1); plt.plot(t/50, q1)
plt.title('$q_1$'); plt.grid()
plt.subplot(2,4,2); plt.plot(t/50, q2)
plt.title('$q_2$'); plt.grid()
plt.subplot(2,4,3); plt.plot(t/50, q3)
plt.title('$q_3$'); plt.grid()
plt.subplot(2,4,4); plt.plot(t/50, q4)
plt.title('$q_4$'); plt.grid()
plt.subplot(2,4,5); plt.plot(t/50, q5)
plt.title('$q_5$'); plt.grid()
plt.subplot(2,4,6); plt.plot(t/50, q6)
plt.title('$q_6$'); plt.grid()
plt.subplot(2,4,7); plt.plot(t/50, q7)
plt.title('$q_7$'); plt.grid()
plt.subplot(2,4,8); plt.plot(t/50, q8)
plt.title('$q_8$'); plt.grid()
```
