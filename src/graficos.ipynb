```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
xcur = []
ycur = []
zcur = []

with open("/home/benja/xactual.txt", 'r') as archivo:
    lineas = archivo.readlines()
    for linea in lineas:
        datos = linea.strip().split(' ')  # Suponiendo que los datos estén separados por comas
        xcur.append(float(datos[1]))
        ycur.append(float(datos[2]))
        zcur.append(float(datos[3]))
```

```python
xdes = []
ydes = []
zdes = []

with open("/home/benja/xdeseado.txt", 'r') as archivo:
    lineas = archivo.readlines()
    for linea in lineas:
        datos = linea.strip().split(' ')  # Suponiendo que los datos estén separados por comas
        xdes.append(float(datos[1]))
        ydes.append(float(datos[2]))
        zdes.append(float(datos[3]))
```

```python
q1cur = []
q2cur = []
q3cur = []
q4cur = []
q5cur = []
q6cur = []

with open("/home/benja/qactual.txt", 'r') as archivo:
    lineas = archivo.readlines()
    for linea in lineas:
        datos = linea.strip().split(' ')  # Suponiendo que los datos estén separados por comas
        if len(datos) > 6:  # Verifica que haya al menos 7 elementos
            try:
                q1cur.append(float(datos[1]))
                q2cur.append(float(datos[2]))
                q3cur.append(float(datos[3]))
                q4cur.append(float(datos[4]))
                q5cur.append(float(datos[5]))
                q6cur.append(float(datos[6]))
            except ValueError:
                print(f"Error de conversión en la línea: {linea.strip()}")
        else:
            print(f"Línea ignorada por falta de datos: {linea.strip()}")
```

```python
q1des = []
q2des = []
q3des = []
q4des = []
q5des = []
q6des = []

with open("/home/benja/qdeseado.txt", 'r') as archivo:
    lineas = archivo.readlines()
    for linea in lineas:
        datos = linea.strip().split(' ')  # Suponiendo que los datos estén separados por comas
        if len(datos) > 6:  # Verifica que haya al menos 7 elementos
            try:
                q1des.append(float(datos[1]))
                q2des.append(float(datos[2]))
                q3des.append(float(datos[3]))
                q4des.append(float(datos[4]))
                q5des.append(float(datos[5]))
                q6des.append(float(datos[6]))
            except ValueError:
                print(f"Error de conversión en la línea: {linea.strip()}")
        else:
            print(f"Línea ignorada por falta de datos: {linea.strip()}")
```

```python
ef_amor1 = []
ef_amor2 = []
ef_amor3 = []
ef_amor4 = []
ef_amor5 = []
ef_amor6 = []

with open("/home/benja/amor.txt", 'r') as archivo:
    lineas = archivo.readlines()
    for linea in lineas:
        datos = linea.strip().split(' ')  # Suponiendo que los datos estén separados por comas
        if len(datos) > 6:  # Verifica que haya al menos 7 elementos
            try:
                ef_amor1.append(float(datos[1]))
                ef_amor2.append(float(datos[2]))
                ef_amor3.append(float(datos[3]))
                ef_amor4.append(float(datos[4]))
                ef_amor5.append(float(datos[5]))
                ef_amor6.append(float(datos[6]))
            except ValueError:
                print(f"Error de conversión en la línea: {linea.strip()}")
        else:
            print(f"Línea ignorada por falta de datos: {linea.strip()}")
```
